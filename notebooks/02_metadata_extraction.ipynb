```python
def extract_case_name(text):

    lines = [line.strip() for line in text.split("\n") if line.strip()]

    for i, line in enumerate(lines[:40]):

        if line.lower() == "v.":
            return f"{lines[i-1]} v. {lines[i+1]}"

        if " v." in line.lower():
            return line

    return None
```

```python
def extract_header(text, n_lines=50):

    lines = text.split("\n")

    return "\n".join(lines[:n_lines])
```

```python
def extract_body(text, n_lines=50):

    lines = text.split("\n")

    return "\n".join(lines[n_lines:])
```

### Document Segmentation

A simple line-based segmentation strategy was tested by treating the first 50 lines of each judgment as the document header and the remaining text as the body.

The header consistently contains high-value metadata including:
- Case title
- Date of judgment
- Judges
- Jurisdiction
- Appeal information

The body contains:
- Facts of the case
- Legal arguments
- Statutory references
- Judicial reasoning

This suggests that metadata extraction and content analysis should be treated as separate processing stages.

```python
def extract_year(file):

    parts = file.stem.split("_")

    if parts[0].isdigit():
        return int(parts[0])

    return int(parts[1])
```

```python
import re

date_patterns = [
    r"[A-Z]+\s+\d{1,2},\s+\d{4}",
    r"[A-Z][a-z]+\s+\d{1,2},\s+\d{4}",
    r"\d{1,2}\s+[A-Z][a-z]+\s+\d{4}",
    r"[A-Z]+\d{1,2},\d{4}"
]
```

```python
def extract_date(header):

    for pattern in date_patterns:

        match = re.search(pattern, header)

        if match:
            return match.group()

    return None
```

```python
def extract_judges(header):

    match = re.search(r"\[(.*?)\]", header, re.DOTALL)

    if match:
        return match.group(1)

    return None
```

```python
def extract_case_type(header):

    lines = header.split("\n")

    for line in lines:

        if "JURISDICTION" in line.upper():
            return line.strip()

    return None
```

```python
def extract_metadata(file):

    with open(file, encoding="utf-8") as f:
        text = f.read()

    header = extract_header(text)

    metadata = {
        "file": file.name,
        "year": extract_year(file),
        "date": extract_date(header),
        "judges": extract_judges(header),
        "case_type": extract_case_type(header)
    }

    return metadata
```
